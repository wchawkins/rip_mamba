# Columns to engineer
### Inches--- loc_x and loc_y are in undefined arbitrary units. I want to convert them into inches

### Angle--- For a swish the angle of the shot doesn't matter however for non-swishes the angle at which the shot was taken in relationship to the basebline plays a large role in shooting percentage

### Remaining --- Currently there are minutes/seconds remaining in the period I want to engineer total seconds remaining in the period in which the shot was taken (Completed)




```python
# Determination of loc_x loc_y units for converstion to inches
# loc_x	loc_y	distance
# 167	72	18
a_squared = 167**2
b_squared = 72**2
c_squared = a_squared + b_squared

c_length = math.sqrt(c_squared)
x_y_feet = c_length/18
print(x_y_feet)

# each x/y unit corresponds to .1 feet in any given direction
```

```python
# minutes remaining and seconds remaining should be combined into time remaining... We should engineer that.
x['remaining_time'] = x['minutes_remaining'] * 60 + x['seconds_remaining']
```

```python
# Let's figure out what the loc_x, loc_y, lon, lat mean

nona =  x[pd.notnull(x['shot_made_flag'])]

alpha = 0.02
plot = plt.figure(figsize=(10,10))

# loc_x and loc_y
plt.subplot(121)
plt.scatter(nona.loc_x, nona.loc_y, color='gold', alpha=alpha)
plt.title('loc_x and loc_y')

# lat and lon
plt.subplot(122)
plt.scatter(nona.lon, nona.lat, color='purple', alpha=alpha)
plt.title('lat and lon')

plot.savefig("shot_chart.png")
```
