```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
raw_df = pd.read_csv("data.csv")
```

```python
# There's no missing data
raw_df.describe()
```

```python
# minutes remaining and seconds remaining should be combined into time remaining... We should engineer that.
raw_df['remaining_time'] = raw_df['minutes_remaining'] * 60 + raw_df['seconds_remaining']
```

```python
# Let's figure out what the loc_x, loc_y, lon, lat mean

nona =  raw_df[pd.notnull(raw_df['shot_made_flag'])]

alpha = 0.02
plt.figure(figsize=(10,10))

# loc_x and loc_y
plt.subplot(121)
plt.scatter(nona.loc_x, nona.loc_y, color='blue', alpha=alpha)
plt.title('loc_x and loc_y')

# lat and lon
plt.subplot(122)
plt.scatter(nona.lon, nona.lat, color='green', alpha=alpha)
plt.title('lat and lon')
```
